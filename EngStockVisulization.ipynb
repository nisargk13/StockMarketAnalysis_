```
#For reading data pandas_datareader will be used
import pandas_datareader.data as web
import datetime
```

```
start = datetime.datetime(2020,1,1)
end = datetime.datetime(2020,6,15)
```

```
#Stcok name is GOOGL and source for pricing is yahoo
Enghouse = web.DataReader("ENGH", 'yahoo',start,end)
```

```
#To see each column
Enghouse['Open']
```

```
import matplotlib.pyplot as plt
```

```
#Plot data of Column Open using Plot()
Enghouse['Open'].plot()
plt.show()
```

```
#Plot data of Column Open and Close with Legend
Enghouse['Open'].plot(label = 'Open Price')
Enghouse['Close'].plot(label = 'Close Price')
plt.legend()
plt.show()
```

```
#Plot data of Column with Legend, Title, lable and size of 15,7

Enghouse['Open'].plot(label = 'Open Price',figsize=(15,7))
Enghouse['Close'].plot(label = 'Close Price')
Enghouse['High'].plot(label = 'High Price')
Enghouse['Low'].plot(label = 'Low Price')
plt.title('EngHouse Stock Price')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
